# Stochastic optimisation on TSP

```python
from processing import read_dataset, draw_way, launching, generate_new_dataset
from searches import monte_carlo, random_walk, hill_climb, annealing, genetic_algo, total_l1
```

#### 0. Generate dataset
First, we have to generate random dataset (if we don't have one).

```python
datafile = generate_new_dataset(100)
data = read_dataset(datafile)
draw_way(data, 'Input', total_l1(data))
```

### Launching algorithms
Now, let's run our algorithms with different parameters and look at the results.

#### 1. Monte Carlo

```python
launching(datafile, monte_carlo, 'Monte Carlo', 100000)
launching(datafile, monte_carlo, 'Monte Carlo', 1000000)
```

#### 2. Random walk

```python
launching(datafile, random_walk, 'Random walk', 100000)
launching(datafile, random_walk, 'Random walk', 1000000)
```

#### 3. Hill climb

```python
launching(datafile, hill_climb, 'Hill climb', 40)
launching(datafile, hill_climb, 'Hill climb', 400)
```

#### 4. Simulated annealing

```python
launching(datafile, annealing, 'Simulated annealing', 5, delta_t = 0.01, t_max = 500)
launching(datafile, annealing, 'Simulated annealing', 50, delta_t = 0.01, t_max = 500)
```

#### 5. Genetic algorithm

```python
launching(datafile, genetic_algo, 'Genetic algorithm', 100, size = 500, mut_rate = 0.1, sel_rate = 0.08)
launching(datafile, genetic_algo, 'Genetic algorithm', 1000, size = 500, mut_rate = 0.1, sel_rate = 0.08)
```
